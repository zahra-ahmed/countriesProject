```python
# dependencies

import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
# connect to local database

rds_connection_string = "usr1:password@localhost:5432/countries_project"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# check for tables

engine.table_names()
```

```python
country_info = pd.read_sql('select country_names.iso_code, country_names.country, country_info.region,\
country_names.capital, country_info.income_group from country_info inner join country_names\
 on country_info.iso_code=country_names.iso_code', con=engine)
country_info.head()
```

```python
country_info.to_json(orient='records')
```

```python
with open('country_info.json', 'w', encoding='utf-8') as file:
    country_info.to_json(file, force_ascii=False, orient='records')
```
